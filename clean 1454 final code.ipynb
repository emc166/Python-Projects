```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

NOAA = pd.read_csv('Baton_Rouge_NOAA.csv')  # load data set
x = NOAA.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
y = NOAA.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
model = LinearRegression()  # create object for the class
model.fit(x, y)  # perform linear regression

print("Model Slope:", model.coef_[0])
print("Model Intercept:", model.intercept_)
r_sq = model.score(x, y)
print('Coefficient Of Determination:', r_sq)

plt.scatter(x, y)
plt.plot(x, y, color='blue')
plt.title('Avergae Temperature of Baton Rouge Over A Course Of One Month (May) From 1970 - 2007')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm

NOAA = pd.read_csv('Baton_Rouge_NOAA.csv')  # load data set
x = NOAA.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
y = NOAA.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
x = sm.add_constant(x)

model = sm.OLS(y, x)

results = model.fit()

print(results.summary())

print('coefficient of determination:', results.rsquared) # .rsquared holds .

print('adjusted coefficient of determination:', results.rsquared_adj) # .rsquared_adj represents adjusted  ( corrected according to the number of input features).

print('regression coefficients:', results.params) # .params refers the array with , , and  respectively.

```
